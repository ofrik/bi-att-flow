```python
import json
import pandas as pd
import os
from tqdm import tqdm
```

```python
SQUAD_TRAIN_DATA = "data/squad/train-v1.1.json"
```

```python
with open(SQUAD_TRAIN_DATA,'r') as f:
    squad = json.load(f)
```

```python
squad_data = squad["data"]
```

```python
df_data = {"id":[],"title":[],"p":[],"q":[],"a":[]}
for subject in tqdm(squad_data):
    title = subject["title"]
    for paragraph in subject["paragraphs"]:
        p = paragraph["context"]
        items = paragraph["qas"]
        for item in items:
            idn = item["id"]
            q = item["question"]
            a = item["answers"][0]["text"]
            if len(item["answers"])>1:
                print("more than one answer")
            df_data["id"].append(idn)
            df_data["title"].append(title)
            df_data["p"].append(p)
            df_data["q"].append(q)
            df_data["a"].append(a)
```

```python
df = pd.DataFrame(df_data)
```

```python
df.to_csv("data/squad/processed_data.csv",index=False,encoding="UTF8")
```
